```python
import plotly.graph_objects as go
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

x, y = np.meshgrid(x, y)

z = x**2 * y

fig = go.Figure(data=[go.Surface(z=z, x=x, y=y)])

fig.update_layout(
    title='3D Plot of the function f(x,y) = x^2y', 
    autosize=False,
    width=500,
    height=500,
    margin=dict(l=65, r=50, b=65, t=90)
)

fig.show()
```

```python
# Create a Set of Points that Lie on the Circle

theta = np.linspace(0, 2*np.pi, 100)

x = np.cos(theta)
y = np.sin(theta)

# Create the Plot
fig = go.Figure(data=go.Scatter(x=x, y=y, mode='lines'))

# Set the Aspect Ratio to Equal so the Circle does not look like an Ellipse
fig.update_yaxes(
    scaleanchor = 'x',
    scaleratio = 1 
)

fig.show()
```

```python
# Create a Grid for the Function
x = np.linspace(-1.5, 1.5, 100)
y = np.linspace(-1.5, 1.5, 100)

x, y = np.meshgrid(x, y)

z = x**2 * y

# Create a Set of Points that Lie on the Circle
theta = np.linspace(0, 2*np.pi, 100)

x_circle = np.cos(theta)
y_circle = np.sin(theta)
z_circle = x_circle**2 * y_circle

# Create the Surface Plot for the Function
fig = go.Figure(data=[go.Surface(z=z, x=x, y=y, colorscale='Viridis', opacity=0.8)])

# Add the Circle to the Plot
fig.add_trace(go.Scatter3d(x=x_circle, y=y_circle, z=z_circle, mode='lines'))

fig.show()
```

```python
# Create a Grid for the Function
x = np.linspace(-1.5, 1.5, 100)
y = np.linspace(-1.5, 1.5, 100)

x, y = np.meshgrid(x, y)

z = x**2 * y

# Create a Set of Points that Lie on the Circle
theta = np.linspace(0, 2*np.pi, 100)

x_circle = np.cos(theta)
y_circle = np.sin(theta)

# Create the Surface Plot for the Function
fig = go.Figure(data=go.Contour(x=x[0, :], y=y[:, 0], z=z))

# Add the Circle to the Plot
fig.add_trace(go.Scatter(x=x_circle, y=y_circle, mode='lines', line=dict(color='white')))

# Set Aspect Ratio
fig.update_layout(
    autosize=False,
    width=500,
    height=500,
    yaxis=dict(
        scaleanchor = 'x',
        scaleratio = 1 
    )
)

fig.show()
```

```python
# Create a Grid for the Function
x = np.linspace(-2, 2, 10)
y = np.linspace(-2, 2, 10)

x, y = np.meshgrid(x, y)

z = x**2 * y

# Compute the Gradient
U = 2*x*y
V = x**2

plt.figure(figsize=(7,7))

# Draw the Contour Plot
plt.contour(x, y, z, levels=10, cmap='viridis')

# Draw the Gradient vector Field
plt.quiver(x, y, U, V)

plt.title('Contour plot and Gradient Vector Field of $f(x, y) = x^2y$')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```

```python
# Create a Grid for the Function
x = np.linspace(-2, 2, 10)
y = np.linspace(-2, 2, 10)

x, y = np.meshgrid(x, y)

z = x**2 + y**2

# Compute the Gradient
U = 2*x
V = 2*y

plt.figure(figsize=(7,7))

# Draw the Contour Plot
plt.contour(x, y, z, levels=10, cmap='viridis')

# Draw the Gradient vector Field
plt.quiver(x, y, U, V)

plt.title('Contour plot and Gradient Vector Field of $f(x, y) = x^2y$')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```
